# RG Beta Function from QGF Flow

This notebook simulates the running of an effective coupling under categorical renormalization flow in the QGF framework.  
It estimates a beta function using either entropy-weighted scaling or fusion-dimension-driven evolution, as discussed in Section 6.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# RG Flow


t = np.linspace(0, 10, 200)  # RG time
g0 = 0.5                     # initial value

# Example flow: entropic or dimension-weighted decay
g = g0 / (1 + 0.25 * t)      # toy QGF beta flow

# Compute beta function Î²(g) = dg/dt
dg_dt = -0.25 * g**2         # analytic form for this model
```

```python
# Plot RG Flow and Beta Function

fig, ax = plt.subplots(1, 2, figsize=(10, 4))

ax[0].plot(t, g, label="g(t)")
ax[0].set_title("Coupling Flow")
ax[0].set_xlabel("RG Time $t$")
ax[0].set_ylabel("Effective Coupling $g(t)$")
ax[0].grid(True)

ax[1].plot(g, dg_dt, label=r"$\beta(g)$", color="darkred")
ax[1].set_title("Beta Function")
ax[1].set_xlabel("$g$")
ax[1].set_ylabel(r"$\beta(g)$")
ax[1].grid(True)

plt.tight_layout()
plt.show()
```

# Interpretation

This beta function is consistent with an asymptotically free or entropically suppressed flow,  
as expected from QGF categorical structures where effective couplings scale with quantum dimensions or mutual information gradients.
