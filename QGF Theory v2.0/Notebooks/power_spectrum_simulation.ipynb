# Power Spectrum Simulation (QGF Scalar Tilt)

This notebook visualizes the scalar power spectrum derived from categorical QGF dynamics.  
It confirms the expected tilt around \( n_s = 0.964 \) as discussed in Section 7.3 of the main text.


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv("../data/Simulated_Scalar_Power_Spectrum_n_s_0.964.csv")
data.columns = ["k", "P(k)"]

data.head()
```

```python
# Plot Power Spectrum

plt.figure(figsize=(6, 4))
plt.loglog(data["k"], data["P(k)"], marker='o', linewidth=1.2)
plt.title("Simulated Scalar Power Spectrum (QGF)")
plt.xlabel("Mode $k$")
plt.ylabel("Power Spectrum $P(k)$")
plt.grid(True, which="both", ls="--")
plt.tight_layout()
plt.show()
```

```python
# Fit and Estimate Spectral Tilt

log_k = np.log(data["k"])
log_P = np.log(data["P(k)"])

slope, intercept = np.polyfit(log_k, log_P, 1)
n_s = slope + 1

print("Estimated spectral index n_s ", round(n_s, 4))
```

#Interpretation

The log-log fit yields a slope consistent with a scalar tilt \( n_s \approx 0.964 \),  
matching Planck constraints and validating the entanglement-driven inflation model in QGF.
